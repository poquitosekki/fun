```r
library(readr)
library(janitor)
library(ggplot2)
```

```r
dataset <- read_csv('./StudentsPerformance.csv', show_col=cols()) %>%
           clean_names()
head(dataset)
```

```r
summary(dataset)
```

```r
names(dataset)
```

```r
ggplot(data=dataset) +
    geom_bar(mapping=aes(x=race_ethnicity, fill=gender))
```

```r
ggplot(dataset) + 
    geom_boxplot(mapping=aes(x=race_ethnicity,y=math_score,fill=gender))
```

```r
ggplot(data=dataset) +
    geom_bar(mapping=aes(x=parental_level_of_education,fill=race_ethnicity))
```

```r
# The position='fill' option is great when we want to compare proportions without worrying about the height
ggplot(data=dataset) +
    geom_bar(mapping=aes(x=parental_level_of_education,fill=race_ethnicity),position='fill')
```

```r
ggplot(data=dataset) +
    geom_bar(mapping=aes(x=test_preparation_course, fill=gender))
```

```r
names(dataset)
```

```r
ggplot(data=dataset,mapping=aes(x=reading_score,y=writing_score)) +
    geom_point(size=2) +
    geom_smooth(size=2,se=FALSE) +
    theme_minimal()
```

```r
ggplot(data=dataset,mapping=aes(x=reading_score,y=math_score)) +
    geom_point(size=2) +
    geom_smooth(size=2,se=FALSE) +
    theme_minimal()
```

```r
cor(dataset$writing_score, dataset$reading_score)
cor(dataset$math_score, dataset$reading_score)
```

```r
ggplot(data=dataset) +
    geom_bar(mapping=aes(x=test_preparation_course,fill=lunch))
```

```r
ggplot(data=dataset,mapping=aes(x=test_preparation_course,fill=lunch)) +
    geom_bar(position='identity',alpha=1/5)
```

```r
ggplot(data=dataset,mapping=aes(x=test_preparation_course,fill=lunch)) +
    geom_bar(position='fill')
```

```r
ggplot(data=dataset,mapping=aes(x=reading_score,y=writing_score)) +
    geom_point(position='jitter') +
    geom_smooth(se=FALSE)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Computation

```python
# Getting Statistics using various operations from linear algebra
# Vector definitions
v = [4,6,2] # Vector of Variable X
a = [1,1,1]

# Sum of V
vsum = np.dot(v,a)
# Mean of V 
vmean = (vsum/np.dot(a,a))
# Standard Deviation of V
vstd = 0
# Variance of V
vvar = 0

# Show results
print('Sum\tMean\tStd\tVariance')
print(vsum, vmean, vstd, vvar, sep='\t')
```

```python
from statistics import fsum,fmean,stdev,variance
print('Sum\tMean\tStd\tVariance')
print('%.2f\t%.2f\t%.2f\t%.2f' % (fsum(v),fmean(v),stdev(v),variance(v)))
```

# Visualization

```python
# Draw Plot
fig = plt.figure(num=1,clear=True,figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection='3d')

# Draw Plane
point = np.array([0,0,0])
normal = np.array([1,1,1])
d = -point.dot(normal)
x,y = np.meshgrid(range(10),range(10))
z = (-normal[0]*x-normal[1]*y-d)*1./normal[2]
ax.plot_surface(x,y,z,alpha=0.5,color='y')

# Draw Vectors
start = [0,0,0]
X,Y,Z = zip(start, start, start)
ax.quiver(X,Y,Z,v[0],v[1],v[2],color='k')
ax.quiver(X,Y,Z,a[0],a[1],a[2],color='r')
ax.quiver(X,Y,Z,4*a[0],4*a[1],4*a[2])

# Plot Settings
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-5,5])
ax.set_ylim([-5,10])
ax.set_zlim([-5,5])
ax.view_init(10,20)
```
